### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query_1 = {"scores.Hygiene": {"$eq": 20}}
fields_1 = {"_id": 0, "BusinessName": 1, "scores.Hygiene": 1}

# Use count_documents to display the number of documents in the result
result_count_1 = establishments.count_documents(query_1)
print(result_count_1)

# Display the first document in the results using pprint
result_doc_1 = establishments.find_one(query_1)
pprint(result_doc_1)
```

```python
# Convert the result to a Pandas DataFrame
result_1 = establishments.find(query_1, fields_1)
df_1 = pd.DataFrame(result_1)

# Display the number of rows in the DataFrame
print(len(df_1))

# Display the first 10 rows of the DataFrame
print(df_1.head(10))
```

```python
result_1 = establishments.find(query_1, fields_1)
print(list(result_1))
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with City of London Corporation as the Local Authority and has a RatingValue greater than or equal to 4.
query_2 = {"$and":[{"LocalAuthorityName": "City of London Corporation"}, {"RatingValue": {"$gte": 4}}]}
fields_2 = {"_id": 0, "BusinessName": 1, "LocalAuthorityName": 1, "RatingValue": 1}

# Use count_documents to display the number of documents in the result
result_count_2 = establishments.count_documents(query_2)
print(result_count_2)

# Display the first document in the results using pprint
result_doc_2 = establishments.find_one(query_2)
pprint(result_doc_2)
```

```python
# Convert the result to a Pandas DataFrame
result_2 = establishments.find(query_2, fields_2)
df_2 = pd.DataFrame(result_2)

# Display the number of rows in the DataFrame
print(len(df_2))

# Display the first 10 rows of the DataFrame
print(df_2.head(10))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = .01
latitude = 51.490142
longitude = 0.08384

query = {
    "RatingValue": 5,
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
}

sort = [("Hygiene", 1)]  # Sort by lowest hygiene score
limit = 5  # Get the top 5 establishments

# Print the results
results_3 = establishments.find(query).sort(sort).limit(limit)
```

```python
# Convert result to Pandas DataFrame
df = pd.DataFrame(results_3)
df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
match_query = {'$match': {"scores.Hygiene": 0}}

group_query = {'$group': 
               {"_id": "$LocalAuthorityName",
                'count': {'$sum': 1},
                }}

sort_values = {'$sort': {'count': -1}}

pipeline = [match_query, group_query, sort_values]

pipe_results = establishments.aggregate(pipeline)

# Print the number of documents in the result
result_list = list(pipe_results)
result_count = len(result_list)
print("Number of documents in result:", result_count)

# Print the first 10 results
pprint(list(result_list)[:10])
```

```python
results = establishments.aggregate(pipeline)

# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(df.shape)

# Display the first 10 rows of the DataFrame
df.head(10)
```
